```python
mutation_df =  pd.read_csv(mutation_f,sep="\t")
mutation_df
```

```python
merge_df = mutation_df
```

```python
# 行索引转为第一列，并命名为 Group
merge_df.reset_index(inplace=True)
merge_df.rename(columns={"index":"Group"},inplace=True)
merge_df
```

```python

def mark_p_value(row):
    if row['p_value'] <= 0.001:
        return "***"
    elif row['p_value'] <= 0.01:
        return "**"
    elif row['p_value'] <= 0.05:
        return "*"
    else:
        return "NS."
 
merge_df['p_value_mark'] = merge_df.apply(mark_p_value, axis=1)
merge_df
```

```python
output_f = os.path.join(output_dir, "20250606_21_MDS_sAML_AML_immune_chisq_cell_kmean_merge_vs_mutation_stat.xls")
merge_df.to_csv(output_f,index = False)
```
