```python
## input
Karyotype_f = "/share/home/wyz/Project/result/2025/202506/20250605_20250520_1_cibersortx_tumor_cell_kmean_5448_MDS_sAML_AML_merge/20250606_9_MDS_sAML_AML_tumor_cell_kmean_and_clinical_data_cor_data/20250606_1_stat_df_chiseq_results_Karyotype.xls"
```

```python
Karyotype_df =  pd.read_csv(Karyotype_f,sep="\t")
Karyotype_df
```

```python
merge_df = Karyotype_df
```

```python
# 行索引转为第一列，并命名为 Group
merge_df.reset_index(inplace=True)
merge_df.rename(columns={"index":"Group"},inplace=True)
merge_df
```

```python
# 根据条件标记新的列
def mark_p_value(row):
    if row['p_value'] <= 0.001:
        return "***"
    elif row['p_value'] <= 0.01:
        return "**"
    elif row['p_value'] <= 0.05:
        return "*"
    else:
        return "NS."
 
merge_df['p_value_mark'] = merge_df.apply(mark_p_value, axis=1)
merge_df
```

```python
output_f = os.path.join(output_dir, "20250606_23_MDS_sAML_AML_tumor_chisq_cell_kmean_merge_vs_Karyotype.xls")
merge_df.to_csv(output_f,index = False)
```
