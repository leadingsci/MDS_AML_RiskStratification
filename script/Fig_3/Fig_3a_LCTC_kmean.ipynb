```r
# k-means聚类
set.seed(123)
best_k = 2
kmeans_result_c2 <- kmeans(cell_type_scaled, centers = best_k, nstart = 25)
print(kmeans_result_c2)
```

```r
# k-means
set.seed(123)
best_k = 5
kmeans_result_c5 <- kmeans(cell_type_scaled, centers = best_k, nstart = 25) ## 
print(kmeans_result_c5)
```

```r
## dot and text
plot_object <- fviz_cluster(
  kmeans_result_c5, 
  data = cell_type_scaled,
  show.clust.cent = TRUE,  
  geom = c("point","text"),          
  ellipse.type = "convex", 
    # ellipse.type = "norm",
  #palette = "jco",         
  palette = c("#ca8282","#66abda","#e19793","#898989","#f5d966"), 
  labelsize = 5,          
  pointsize = 2,
  ggtheme = theme_minimal(),
  main = "Cluster kmean=5",
)

output_f_path = file.path(output_dir,"20250703_15_1_k5_MDS_sAML_sAML_tumor_cell_percent_kmean_split_group_point_text_convex.pdf")
ggsave(filename=output_f_path,plot=plot_object,device=pdf,width=150,height=150,units="mm")
```
